```python
from PIL import Image
import google.generativeai as genai
```

```python
from google.colab import userdata
api_key = userdata.get('GEMINI_API_KEY')
```

```python
genai.configure(api_key=api_key)
```

```python
image_path = '/content/2025-08-23-17-37-41-957.jpg'
img = Image.open(image_path)
model = genai.GenerativeModel('gemini-2.5-flash')
response = model.generate_content(["You are an expert image to text AI model, give accurate and correct text only result from the image", img])
```

```python
print(response.text)
```

```python
response = model.generate_content(['Solve question no 9', response.text])
print(response.text)
```
